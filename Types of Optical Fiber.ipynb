# Types of Optical Fibers

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Assigning Radius of Core & Cladding

```python
a = 50 #radius of core
b = 125 #radius of cladding
```

### Assigning Refractive Index (RI) Values

```python
n1 = 1.457 #RI of core
n2 = 1.45 #RI of cladding
```

## Step Indexed Fiber

```python
r = np.linspace(0, b, b + 1) #radius of the fiber
n = np.linspace(0, b, b + 1) #change in RI
for i in range(len(r)):
    if i <= 50:
        n[i] = n1
    else:
        n[i] = n2
```

```python
#Plotting the Relation between Refractive Index & Radius
plt.figure()
plt.plot(r, n)
plt.xlabel('Variable defining the Optical Fiber Radius (r)')
plt.ylabel('Refractive Index n(r)')
plt.title('Refractive Index Profile\n Step Indexed Fiber')
```

## Graded Indexed Fiber

```python
delta = 0.001
plt.figure()
for alpha in range(1,10):
    for i in range(len(r)):
        if i < a:
            m = 2*delta*((i/a)**alpha)
            n[i] = n1*((1 - m)**0.5)
        else:
            m = 2*delta
            n[i] = n1*((1 - m)**0.5)
    plt.plot(r, n, label = 'Î±={}'.format(alpha))
plt.xlabel('Variable defining the Optical Fiber Radius (r)')
plt.ylabel('Refractive Index n(r)')
plt.title('Refractive Index Profile\n Graded Indexed Fiber')
plt.legend()
```
