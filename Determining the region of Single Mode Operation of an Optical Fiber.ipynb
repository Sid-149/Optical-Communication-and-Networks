# Determining the region of Single Mode Operation of an Optical Fiber

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#assigning refractive index (RI)
n1 = 1.46 #RI of core
n2 = 1.455 #RI of cladding
```

```python
#calculating numerical aperture (NA)
NA = (n1**2 - n2**2)**0.5
print('NA = {}'.format(NA))
```

```python
#calculating V number
d = 8e-06 #diameter
lambda1 = np.arange(0.8, 1.6, 0.01)
vno = np.arange(0.8, 1.6, 0.01)

for i in range(len(lambda1)):
    vno[i] = (np.pi*d*NA)/(lambda1[i]*1e-06)
```

```python
#plotting
plt.figure(figsize = (10,6))
plt.plot(lambda1, vno, 'r')
plt.xlabel('Wavelength (in micrometers)')
plt.ylabel('V no.')
plt.title('Wavelength v/s V no. for d = 8 Î¼m')
plt.axhspan(min(vno), 2.405, color = 'black', alpha = 0.5)
plt.axvspan(1.26, 1.6, color = 'yellow', alpha = 0.5)
plt.annotate('Single Mode Operation', xy = (1.29, 2))
```
