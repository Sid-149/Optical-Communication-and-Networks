# RMS Pulse Broadening due to material dispersion in Multimode Step Index Fiber

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#assigning values
lambda1 = 850
lambda2 = 1300
c = 3e05
sig_lam = np.arange(10, 100, 10)
sig_m1 = np.float16(np.arange(10, 100, 10))
#sig_m1 = sig_m1.astype('float16')
sig_m2 = np.float16(np.arange(10, 100, 10))
#sig_m2 = sig_m2.astype('float16')
factor = 0.025
```

```python
#calculation
M1 = factor/(c*lambda1)
M2 = factor/(c*lambda2)

for i in range(len(sig_lam)):
    sig_m1[i] = M1*sig_lam[i]*1e12
    sig_m2[i] = M2*sig_lam[i]*1e12
```

```python
#plotting
plt.figure()
plt.plot(sig_lam, sig_m1)
plt.plot(sig_lam, sig_m2)
plt.xlabel('Spectral Width of Ligth sources (nm)---->')
plt.ylabel('RMS Pulse broadening [ps/(nm-km)]---->')
plt.title('RMS Pulse Broadening \n due to Material Dispersion')
plt.grid()
plt.legend(['Peak Wavelength = 850 nm', 'Peak Wavelength = 1300 nm'], loc = 'upper left')
```
