# Characteristics Curve of LED

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#assigning values
h = 6.26e-34
c = 3e8
lam = 870e-9
tr = 60e-9
tnr = 100e-9
q = 1.6e-19
```

```python
#calculation
t = (tr*tnr)/(tr+tnr)
print('Time: '+str(t)+' seconds')

nint = t/tr
print('NINT:',nint)

I = np.arange(0, 40, 1)*1e-3
pint = [0]*I

for i in range(40):
    pint[i] = (nint*I[i]*h*c)/(q*lam)
```

```python
#plotting
plt.figure()
plt.plot(I, pint)
plt.title('Characteristics Curve of LED')
plt.xlabel('Current(A)----------->')
plt.ylabel('Internally Generated Optical Power(W)----------->')
```
